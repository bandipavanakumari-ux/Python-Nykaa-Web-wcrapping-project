```python
def get_page_contents(url):
    
    '''This function takes url as input and return html page.'''
    
    response= requests.get(url)
    if response.status_code==200:
        doc= bs(response.text,'html.parser')
        return doc
```

```python
def get_productname(doc):

    ''' This function takes html page as input and extract each Product Name '''
    
    productname = []

    for name in doc.find_all('div',{'class':'css-xrzmfa'}):
        productname.append(name.text)
    return productname
```

```python
def get_productprice(doc):
    
    ''' This function take html page as input and extract Product's Origional Price '''
    
    orgprice = [None if i.find('span',class_='css-17x46n5').text == 'MRP:' else i.find('span',class_='css-17x46n5').text
                for i in doc.find_all('div',class_='css-1d0jf8e') if i.find('span',class_='css-17x46n5')]
    
    return orgprice
```

```python
def get_offer_price(doc):
    
    ''' This function take html page as input and extract Product's Offer Price'''
    
    offerprice= [price.text if price else None for price in doc.find_all('span',{'class':'css-111z9ua'})]
    
    return offerprice
```

```python
def getoffer(doc):
    
    ''' This function take html page as imput and extracts Offer % on Product'''
    
    offer = [i.find('span', class_ = 'css-cjd9an').text if i.find('span', class_ = 'css-cjd9an') else None 
             for i in doc.find_all('div', class_='css-1rd7vky')]
    
    return offer    
```

```python
def getreview(doc):
    
    ''' This function take html page as input and extracts Product's number of reviews '''
    
    review = [r.text if r else None for r in doc.find_all('span',{'class':'css-1qbvrhp'})]
    return review
```

```python
def dictionary(productname,orgprice,offerprice,offer,review):

    ''' This function takes different tags and creats List of Dictionary  '''
    
    d = []
    for productname,origionalprice,offerprice,offer,review in zip(productname,orgprice,offerprice,offer,review):
        d.append({'Product Name':productname,
                  'Origional Price':origionalprice,
                  'Offer Price':offerprice,
                  'Offer':offer,
                  'Review':review})
        
        return d
```

```python
def create_df(*args):
    
    '''This function create dataframe from dictionary'''

    return pd.DataFrame(*args)
```

```python
baseurl = 'https://www.nykaa.com/skin/c/8377?page_no='

finaldict = []

for i in range(1,31):
    doc = get_page_contents(baseurl+str(i))
    productname = get_productname(doc)
    orgprice = get_productprice(doc)
    offerprice = get_offer_price(doc)
    offer = getoffer(doc)
    review = getreview(doc)
    
    finaldict += dictionary(productname,orgprice,offerprice,offer,review)
    
df = create_df(finaldict)
df
```

```python
import requests
import pandas as pd
import time

# Base API URL
base_api = "https://www.nykaa.com/gludo/products/list"

# Parameters for the request
params = {
    "pro": "false",
    "filter_format": "v2",
    "app_version": "null",
    "client": "react",
    "root": "nav_3",
    "category_id": "8377",  # Skin category
    "page_no": 1
}

# Empty list to store all products
all_products = []

# Function to extract required fields from product
def extract_product_info(product):
    return {
        "Product Name": product.get("name"),
        "Brand": product.get("brand_name"),
       # "Original Price": product.get("original_price"),
        "Final Price": product.get("final_price"),
        "Discount": product.get("discount"),
        "Rating": product.get("rating"),
        #"Reviews Count": product.get("wishlist_count"),
      #  "Product URL": f"https://www.nykaa.com{product.get('url')}"
    }

# Scrape only pages 1 to 30
with requests.Session() as sess:
    for page in range(1, 41):  # 1 to 30 inclusive
        print(f"Scraping page {page}/41...")
        params["page_no"] = page
        res = sess.get(base_api, params=params)
        products = res.json()["response"]["products"]
        
        for prod in products:
            all_products.append(extract_product_info(prod))
        
        time.sleep(1)  # polite pause

# Create DataFrame
df = pd.DataFrame(all_products)

# Show preview
print(df.head())

# Save to CSV
df.to_csv("Nykaa_skin_care.csv", index=False)
```
